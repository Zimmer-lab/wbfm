```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.neighbors import NearestNeighbors
from wbfm.utils.projects.finished_project_data import ProjectData
```

```python
fname = "/scratch/zimmer/Charles/dlc_stacks/worm3-2021_09_29/project_config.yaml"
project_data = ProjectData.load_final_project_data_from_config(fname)
```

```python
i_time = 0
max_dist = 1.0

zxy_from_seg = project_data.get_centroids_as_numpy(i_time)

df = project_data.final_tracks.copy()
all_neurons = list(df.columns.levels[0])
coords = ['z', 'x', 'y']
zxy_from_tracks = np.array([df[n][coords].iloc[i_time].to_numpy() for n in all_neurons])
```

```python
%%time
algorithm = 'brute'
neighbors_of_seg = NearestNeighbors(n_neighbors=1, radius=max_dist, algorithm=algorithm).fit(zxy_from_seg)

all_matches = [(i, neighbors_of_seg.kneighbors([zxy])[1][0][0]) for i, zxy in enumerate(zxy_from_tracks)]
```

```python
%%time
algorithm = 'kd_tree'
neighbors_of_seg = NearestNeighbors(n_neighbors=1, radius=max_dist, algorithm=algorithm).fit(zxy_from_seg)

all_matches = [(i, neighbors_of_seg.kneighbors([zxy])) for i, zxy in enumerate(zxy_from_tracks)]
```

```python
%%time
algorithm = 'ball_tree'
neighbors_of_seg = NearestNeighbors(n_neighbors=1, radius=max_dist, algorithm=algorithm).fit(zxy_from_seg)

all_matches = [(i, neighbors_of_seg.kneighbors([zxy])) for i, zxy in enumerate(zxy_from_tracks)]
```
