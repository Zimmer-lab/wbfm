```python
import pandas as pd
from tqdm.auto import tqdm

fname = r"C:\dlc_stacks\Charlie-worm3-new-seg\4-traces\green_traces.h5"
green_traces = pd.read_hdf(fname)

fname = r"C:\dlc_stacks\Charlie-worm3-new-seg\4-traces\red_traces.h5"
red_traces = pd.read_hdf(fname)

dlc_fname = r"C:\dlc_stacks\Charlie-worm3-new-seg\3-tracking\postprocessing\combined_3d_tracks.h5"
dlc_tracks = pd.read_hdf(dlc_fname)
```

```python
dlc_tracks
```

```python
# Correct mismatches in DLC-segmentation

old_dlc_names = list(dlc_tracks.columns.levels[0])
new_neuron_names = [f"neuron{i + 1}" for i in range(len(old_dlc_names))]
dlc_name_mapping = dict(zip(new_neuron_names, old_dlc_names))

def _get_dlc_zxy_one_neuron_all_t(new_name):
    old_name = dlc_name_mapping[new_name]
    coords = ['z', 'y', 'x']
    all_dlc_zxy = np.asarray(dlc_tracks[old_name][coords])
    return all_dlc_zxy
```

```python
for name in tqdm(new_neuron_names):
    zxy = _get_dlc_zxy_one_neuron_all_t(name)
    
    red_traces[(name, 'z_dlc')] = zxy[:,0]
    red_traces[(name, 'x_dlc')] = zxy[:,1]
    red_traces[(name, 'y_dlc')] = zxy[:,2]
    
    green_traces[(name, 'z_dlc')] = zxy[:,0]
    green_traces[(name, 'x_dlc')] = zxy[:,1]
    green_traces[(name, 'y_dlc')] = zxy[:,2]
```

```python
# Do training data frame selection from DLC style dataframe
import pandas as pd
import numpy as np
from tqdm.auto import tqdm

fname = r"C:\dlc_stacks\Charlie-worm3-new-seg\2-training_data\all_tracklets.h5"
df = pd.read_hdf(fname)

num_frames = 30
```

```python
# Roll across windows and check how many columns do not have nan

sz = df.shape
# Only need one column from each neuron
missing_vals = np.isnan(df.values)[:, range(0, sz[1], 4)]

num_missing_per_window = {}
for i in tqdm(range(sz[0] - num_frames)):
    start, stop = i, i + num_frames + 1
    # def check_any_nan_in_column(col):
    #     return any(col)

    have_missing = np.apply_along_axis(any, 0, missing_vals[start:stop, :])
    num_missing_per_window[(start, stop)] = sum(~have_missing)
```
