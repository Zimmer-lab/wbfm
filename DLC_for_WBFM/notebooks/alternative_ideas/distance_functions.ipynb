```python
from DLC_for_WBFM.utils.general.distance_functions import confidence_using_tracklet_lengths
import matplotlib.pyplot as plt
import numpy as np
import itertools
from mpl_toolkits.mplot3d import Axes3D
```

```python
lens = np.linspace(1, 100, 100)
inliers = np.linspace(0, 1, 100)
Z0 = np.zeros((100, 100))

for i, j in itertools.product(range(100), range(100)):
    Z0[i, j] = confidence_using_tracklet_lengths(lens[i], inliers[j], gamma_inliers=0.3)
    
Z1 = np.zeros((100, 100))

for i, j in itertools.product(range(100), range(100)):
    Z1[i, j] = confidence_using_tracklet_lengths(lens[i], inliers[j], gamma_inliers=1)
```

```python
X, Y = np.meshgrid(lens, inliers)
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
plt.xlabel("length")
plt.ylabel("percent inliers")

surf = ax.plot_surface(X, Y, Z0,
                       linewidth=0, antialiased=False)
# surf = ax.plot_surface(X, Y, Z1,
#                        linewidth=0, antialiased=False)
```
