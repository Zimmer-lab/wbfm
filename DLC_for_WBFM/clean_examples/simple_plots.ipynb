# Plot some basic traces

```python
neuron_to_plot = 'neuron_062'
```

```python
plt.figure(figsize=(15,5))
project_data.red_traces[neuron_to_plot]['intensity_image'].plot()
plt.title("Red brightness")
```

```python
plt.figure(figsize=(15,5))
y = project_data.green_traces[neuron_to_plot]['intensity_image']
vol = project_data.green_traces[neuron_to_plot]['area']
plt.plot(y)
plt.title("Green raw brightness")
```

```python
plt.figure(figsize=(15,5))
plt.plot(y/vol)
plt.title("Green mean brightness over pixels")
```

```python
plt.figure(figsize=(15,5))
y = project_data.green_traces[neuron_to_plot]['intensity_image']
vol = project_data.green_traces[neuron_to_plot]['area']
plt.plot(y - vol*14)
plt.title("Green integrated and background subtracted brightness")
```

```python
plt.figure(figsize=(15,5))
y1 = project_data.green_traces[neuron_to_plot]['intensity_image']
y2 = project_data.red_traces[neuron_to_plot]['intensity_image']
vol = project_data.green_traces[neuron_to_plot]['area']
plt.plot(y1 / y2)
plt.title("Ratio of raw brightness")
```

```python
plt.figure(figsize=(15,5))
y1 = project_data.green_traces[neuron_to_plot]['intensity_image']
y2 = project_data.red_traces[neuron_to_plot]['intensity_image']
vol = project_data.green_traces[neuron_to_plot]['area']
plt.plot((y1 - vol*14) / (y2 - vol*14))
plt.title("Ratio of background subtracted brightness")
```

# Plot some metadata along with the brightness

```python
plt.figure(figsize=(15,5))
y2 = project_data.red_traces[neuron_to_plot]['intensity_image']
vol = project_data.red_traces[neuron_to_plot]['area']
z = project_data.red_traces[neuron_to_plot]['z']

y = (y2 - vol*14)

plt.plot(y / np.max(y), label='brightness')
plt.plot(z / np.max(z), label='z')

plt.title("Red and z")
plt.legend()
```

```python
plt.figure(figsize=(25,5))
y2 = project_data.red_traces[neuron_to_plot]['intensity_image']
y1 = project_data.green_traces[neuron_to_plot]['intensity_image']
vol = project_data.red_traces[neuron_to_plot]['area']
z = project_data.red_traces[neuron_to_plot]['z']

# y = (y2 - vol*14)
y = y1

plt.plot(y / np.max(y), label='brightness')
# plt.plot(z / np.max(z), label='z')
plt.plot(vol / np.max(vol), label='vol')

plt.title("Red and z")
plt.legend()
```
